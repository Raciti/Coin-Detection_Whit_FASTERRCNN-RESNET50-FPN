#Scrittura CSV


```python
import csv
```

```python
def findIndex(i):
  name = imgs[i].split(".")[0]
  name_m = name + "_mask.png"
  index_m = Masks.index(name_m)
  name_b =  name + "_box.pt"
  index_b = box_dir.index(name_b)
  return index_m, index_b
```

```python
if (len(Masks) != len(imgs)) or len(box_dir) != len(imgs):
    print("Le directory non hanno lo stesso numero di elementi")
else:
  Data = []
  labels = torch.tensor([1])
  for i in range(len(Masks)):
    #print(i+1, "/", len(Masks))
    data = []
    _ , index_b = findIndex(i)
    box_path = boxes_path + box_dir[index_b]
    box = torch.load(box_path)
    """print(imgs[i])
    print(labels)
    print(box)"""
    data = [[imgs[i]] + [labels.tolist()] + [box.squeeze().tolist()]]

    Data.append(data)
    
print(len(Data))
print(Data[0])
```

```python
csv_file = "/content/drive/MyDrive/Colab_Notebooks/Machine_Learning/Progetto/Dataset/dataset.csv"
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    for riga in Data:
      writer.writerows(riga)

print(f"Dati salvati nel file CSV '{csv_file}' con successo.")
```

#Lettura CSV

```python
import ast
import torch
# Percorso del file CSV da aprire
csv_file = "/content/drive/MyDrive/Colab_Notebooks/Machine_Learning/Progetto/Dataset/dataset.csv"
# Lettura dei dati dal file CSV
with open(csv_file, mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
      #print(list(row))
      stringa = row[0]
      tensore_int = torch.tensor(ast.literal_eval(row[1]))
      tensore_float = torch.tensor(ast.literal_eval(row[2])).unsqueeze(0)

      print("Nome Moneta:", stringa)
      print("Labels:", tensore_int)
      print("Box:", tensore_float)

  
```
